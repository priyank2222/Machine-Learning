```python
from keras.layers.core import Dense, Activation, Dropout
from keras.layers.recurrent import LSTM
from keras.models import Sequential
from keras.callbacks import ModelCheckpoint
import lstm, time #helper libraries
import os
```

```python
DATA_DIRECTORY = './all_normallized'
FILE = 'all_60MIN_normalized'
NO_OF_RECORDS = 8
EPOCHS = 15
LOGDIR_ROOT = './logdir/20170502'
checkpoint_every = 1

def get_default_logdir(logdir_root):
    logdir = os.path.join(logdir_root, 'train', STARTED_DATESTRING)
    return logdir


def validate_directories():
    logdir_root = LOGDIR_ROOT
    logdir = logdir_root
    restore_from = logdir

    return {
        'logdir': logdir,
        'logdir_root': logdir_root,
        'restore_from': restore_from
    }
```

```python
directories = validate_directories()
logdir = directories['logdir']
restore_from = directories['restore_from']
```

```python
#Step 1 Load Data
X_train, y_train, X_test, y_test = lstm.load_data(DATA_DIRECTORY + '/' + FILE + '.csv', NO_OF_RECORDS, True)
```

```python
#Step 2 Build Model
model = Sequential()

model.add(LSTM(
    input_dim=1,
    output_dim=50,
    return_sequences=True))
model.add(Dropout(0.2))

model.add(LSTM(
    100,
    return_sequences=False))
model.add(Dropout(0.2))

model.add(Dense(
    output_dim=1))
model.add(Activation('linear'))

start = time.time()
model.compile(loss='mse', optimizer='rmsprop')
print ('compilation time : ' + str(time.time() - start))
```

```python
#Step 3 Train the model
filepath = LOGDIR_ROOT+'/model_{epoch:02d}_{val_loss:.8f}.hdf5'
checkpointer = ModelCheckpoint(filepath, monitor='val_loss', verbose=0, save_best_only=False, save_weights_only=False, mode='auto', period=1)

model.fit(
    X_train,
    y_train,
    batch_size=100,
    nb_epoch=EPOCHS,
    validation_split=0.05, 
    callbacks=[checkpointer])
```

```python
#Step 4 - Plot the predictions!
predictions = lstm.predict_sequences_multiple(model, X_test[:100], NO_OF_RECORDS, NO_OF_RECORDS)
```

```python
lstm.plot_results_multiple(predictions, y_test[:100], NO_OF_RECORDS)
```
